
### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
import seaborn as sns
import matplotlib.pyplot as mlt
```

```python
df['Age'].unique()
```

```python
## Write your code here
import pandas as pd
import numpy as np

# 1. Define the logical mapping
# We use the midpoint or a rank to turn the string into a plot-able number
age_mapping = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
    'Prefer not to say': np.nan  # Use NaN to ignore these in the Scatter Plot
}

# 2. Apply the mapping using the .map() method
# This creates a new numeric column while keeping your original data safe
df['Age_Numeric'] = df['Age'].map(age_mapping)
# 3. Clean up
# Scatter plots cannot plot NaN values, so we drop them for this specific analysis
df_clean = df[['Age_Numeric', 'JobSatPoints_6']].dropna()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age_Numeric', y='JobSatPoints_6', data=df_clean, 
            alpha=0.4, color='blue')

# 4. FORMATTING FOR CLIENTS
plt.title('Agentic AI Analysis: Age vs. Job Satisfaction')
plt.xlabel('Age (Mapped Midpoints)')
plt.ylabel('Satisfaction Score (0-100)')
plt.legend()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
df['ConvertedCompYearly'].dtype
```

```python
df['JobSatPoints_6'].unique()
```

```python
df[['ConvertedCompYearly', 'JobSatPoints_6']]
```

```python
## Write your code here
salary_threshold = df['ConvertedCompYearly'].quantile(0.99)
df_filtered = df[df['ConvertedCompYearly'] < salary_threshold]

df_clean1 = df_filtered[['ConvertedCompYearly', 'JobSatPoints_6', 'Age_Numeric']].dropna()
# 1. Take a sample to reduce clutter (e.g., 2000 rows)
df_sample = df_clean1.sample(n=min(len(df_clean1), 1000), random_state=1)

plt.figure(figsize=(16, 10))

# 2. Use scatterplot with better styling
sns.scatterplot(data=df_sample, 
                x='ConvertedCompYearly', 
                y='JobSatPoints_6', 
                size='Age_Numeric', 
                sizes=(20, 400), 
                hue='Age_Numeric', # Color by age to add depth
                palette='magma', 
                alpha=0.4, 
                edgecolor='white', 
                linewidth=0.5)

# 3. Add a trend line WITHOUT the bubbles (using regplot on top)
sns.regplot(data=df_sample, 
            x='ConvertedCompYearly', 
            y='JobSatPoints_6', 
            scatter=False, 
            color='teal', 
            line_kws={'lw': 3})

plt.title('Cleaned Analysis: Compensation vs. Satisfaction by Age', fontsize=16, pad=20)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.3) # Clean gridlines
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
df['LanguageHaveWorkedWith'].describe()
```

```python
languages = df[['LanguageHaveWorkedWith', 'JobSatPoints_6']].dropna()
languages['Language'] = languages['LanguageHaveWorkedWith'].str.split(';')
languages = languages.explode('Language')
```

```python
# Group by Language and count how many satisfaction scores exist for each
popularity_series = languages.groupby('Language')['JobSatPoints_6'].count()
popularity_series
# Convert the Series to a DataFrame and give it a clear name
popularity_df = popularity_series.reset_index()
popularity_df.columns = ['Language', 'Popularity']
popularity_df
```

```python
popularity_series
```

```python
# Group by Language and calculate the average (mean)
satisfaction_series = languages.groupby('Language')['JobSatPoints_6'].mean()

# Convert to a DataFrame and rename the column
satisfaction_df = satisfaction_series.reset_index()
satisfaction_df.columns = ['Language', 'Avg_Satisfaction']
satisfaction_df
```

```python
# Merge the two individual DataFrames into one final 'lang_stats' table
lang_stats = pd.merge(popularity_df, satisfaction_df, on='Language')
```

```python
## Write your code here
# 3. Visualization
plt.figure(figsize=(14, 8))
sns.scatterplot(data=lang_stats, 
                x='Popularity', 
                y='Avg_Satisfaction', 
                hue='Avg_Satisfaction', 
                palette='coolwarm',
                s=150, # Large point size for clarity
                edgecolor='black')
```

#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.

```python
## Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Identify the 99th percentile threshold
salary_limit = df['ConvertedCompYearly'].quantile(0.99)

# Filter for the bottom 99% to keep the chart "neat"
df_filtered = df[df['ConvertedCompYearly'] < salary_limit]
```

```python
# Select columns and drop NaNs
df_plot = df_filtered[['ConvertedCompYearly', 'JobSatPoints_6', 'Employment']].dropna()

# Shorten long Employment strings so the legend is clean
# (e.g., "Independent contractor, freelancer..." becomes "Freelancer")
df_plot['Employment'] = df_plot['Employment'].str.split(',').str[0]
```

```python
import numpy as np
# Add a small random offset (-0.25 to 0.25) to the Y-axis
df_plot['JobSat_Jittered'] = df_plot['JobSatPoints_6'] + np.random.uniform(-0.25, 0.25, len(df_plot))

```

```python
# Sample 2000 points for a "neat" visual
df_sample = df_plot.sample(n=min(len(df_plot), 2000), random_state=42)

plt.figure(figsize=(14, 8))
sns.scatterplot(data=df_sample, 
                x='ConvertedCompYearly', 
                y='JobSat_Jittered', 
                hue='Employment', 
                style='Employment', # Differentiate with both color AND markers
                alpha=0.6, 
                s=80, 
                palette='viridis')

plt.title('Compensation vs. Job Satisfaction by Employment Type', fontsize=15)
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction (Jittered 1-6 Scale)')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
import pandas as pd
import numpy as np

# 1. Map 'YearsCodePro' strings to numbers
# (Using .replace to keep existing numbers intact)
years_map = {'Less than 1 year': 0.5, 'More than 50 years': 55}
df['YearsCodePro_Num'] = pd.to_numeric(df['YearsCodePro'].replace(years_map), errors='coerce')

# 2. Select specific columns and drop missing values
df_step1 = df[['YearsCodePro_Num', 'Age', 'Country']].dropna()

# 3. Filter for the Top 5 most frequent countries to avoid a "clumsy" legend
top_countries = df_step1['Country'].value_counts().nlargest(5).index
df_step1 = df_step1[df_step1['Country'].isin(top_countries)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define the logical order for Age groups so the X-axis is sorted
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old', 
             '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']

plt.figure(figsize=(14, 8))

# 1. Sort the DataFrame by our 'Age' category using the Categorical order we set earlier
# This ensures the data is physically in the correct order before plotting
df_sorted = df_step1.sort_values(by='Age')

# 2. Re-run the plot using the sorted data
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df_sorted, 
                x='Age', 
                y='YearsCodePro_Num', 
                hue='Country', 
                style='Age', 
                s=100)

plt.xticks(rotation=45)
plt.title('Corrected Order: Experience vs. Age')
plt.show()
```
